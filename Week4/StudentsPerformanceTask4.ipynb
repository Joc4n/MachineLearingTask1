```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Memuat data dari file CSV
file_path = '/content/StudentsPerformance.csv'
data = pd.read_csv(file_path)

# Menambahkan kolom rata-rata skor
data['average score'] = data[['math score', 'reading score', 'writing score']].mean(axis=1)
```

```python
# Analisis rata-rata skor berdasarkan jenis makan siang
lunch_performance = data.groupby('lunch')['average score'].mean().reset_index()

# Visualisasi rata-rata skor berdasarkan jenis makan siang
plt.figure(figsize=(8, 5))
sns.barplot(data=lunch_performance, x='lunch', y='average score', palette='pastel')
plt.title('Rata-rata Skor Berdasarkan Jenis Makan Siang', fontsize=14)
plt.xlabel('Jenis Makan Siang', fontsize=12)
plt.ylabel('Rata-rata Skor', fontsize=12)
plt.show()
```

Kita menganalisis hubungan antara jenis makan siang yang diterima siswa (standard atau free/reduced) dan rata-rata skor mereka. Rata-rata skor dihitung untuk setiap kategori makan siang, dan hasilnya divisualisasikan menggunakan grafik batang (bar plot).

```python
# Analisis rata-rata skor berdasarkan tingkat pendidikan orang tua
education_performance = data.groupby('parental level of education')['average score'].mean().sort_values(ascending=False).reset_index()

# Visualisasi rata-rata skor berdasarkan tingkat pendidikan orang tua
plt.figure(figsize=(10, 6))
sns.barplot(data=education_performance, x='average score', y='parental level of education', palette='coolwarm')
plt.title('Rata-rata Skor Berdasarkan Tingkat Pendidikan Orang Tua', fontsize=14)
plt.xlabel('Rata-rata Skor', fontsize=12)
plt.ylabel('Tingkat Pendidikan Orang Tua', fontsize=12)
plt.show()
```

Analisis ini bertujuan untuk memahami bagaimana tingkat pendidikan orang tua memengaruhi rata-rata skor siswa. Data dikelompokkan berdasarkan tingkat pendidikan orang tua, dan rata-rata skor dihitung untuk setiap kelompok. Hasilnya divisualisasikan dalam bentuk grafik batang dengan sumbu vertikal menunjukkan rata-rata skor dan sumbu horizontal menunjukkan tingkat pendidikan orang tua.



```python
# Distribusi skor berdasarkan gender
plt.figure(figsize=(12, 6))
for i, subject in enumerate(['math score', 'reading score', 'writing score'], 1):
    plt.subplot(1, 3, i)
    sns.boxplot(data=data, x='gender', y=subject, palette='Set2')
    plt.title(f'Distribusi {subject.capitalize()} Berdasarkan Gender', fontsize=12)
    plt.xlabel('Gender', fontsize=10)
    plt.ylabel(subject.capitalize(), fontsize=10)
plt.tight_layout()
plt.show()
```

Langkah ini membandingkan distribusi skor matematika, membaca, dan menulis berdasarkan jenis kelamin siswa (laki-laki atau perempuan). Visualisasi menggunakan box plot untuk menampilkan rentang skor, nilai tengah, dan pencilan (outliers) pada setiap mata pelajaran.

```python
# Pengaruh persiapan ujian terhadap rata-rata skor
prep_performance = data.groupby('test preparation course')['average score'].mean().reset_index()

# Visualisasi pengaruh persiapan ujian
plt.figure(figsize=(8, 5))
sns.barplot(data=prep_performance, x='test preparation course', y='average score', palette='cool')
plt.title('Pengaruh Persiapan Ujian terhadap Rata-rata Skor', fontsize=14)
plt.xlabel('Persiapan Ujian', fontsize=12)
plt.ylabel('Rata-rata Skor', fontsize=12)
plt.show()
```

Langkah terakhir menganalisis pengaruh persiapan ujian terhadap rata-rata skor siswa. Data dikelompokkan berdasarkan status persiapan ujian (completed atau none), dan rata-rata skor dihitung untuk setiap kelompok. Hasilnya divisualisasikan dalam bentuk grafik batang untuk mempermudah perbandingan.

```python
# Hubungan antara skor matematika, membaca, dan menulis
numeric_data = data[['math score', 'reading score', 'writing score']]
sns.pairplot(numeric_data, diag_kind='kde', corner=True, plot_kws={'alpha': 0.6})
plt.suptitle('Hubungan Antara Skor Matematika, Membaca, dan Menulis', y=1.02, fontsize=14)
plt.show()
```

Di sini kita mengeksplorasi hubungan antara skor matematika, membaca, dan menulis untuk menemukan pola atau korelasi di antara ketiga mata pelajaran. Visualisasi menggunakan pair plot, yang menampilkan grafik scatter di antara pasangan variabel, serta distribusi univariat pada diagonal.
